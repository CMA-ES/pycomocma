```python
%pylab nbagg
import cma, como
import numpy as np
```

```python
dimension = 10
num_kernels = 10 # number of single-objective solvers (population size)
```

```python
refpoint = [11, 11]
sigma0 = 0.2
fun1 = cma.ff.sphere
fun2 = lambda x: cma.ff.sphere(x-1)
fun = como.FitFun(fun1, fun2)
```

```python
list_of_solvers = como.get_cmas(-5 + 10 * np.random.rand(num_kernels, dimension), sigma0, inopts={'popsize':32}) # produce `num_kernels cma instances`
moes = como.Sofomore(list_of_solvers, opts = {'archive': True}, reference_point = refpoint) # create a como-cma-es instance
```

```python
while not moes.stop():
    X = moes.ask()
    values = [fun(x) for x in X]
    moes.tell(X, values)
```

```python
moes.logger.plot_front()
```
